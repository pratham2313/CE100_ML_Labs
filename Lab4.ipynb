[2]. Write a function to print largest number from provided 3 values as parameter.

```python
def fun(x,y,z):
  if z>=x and z>=y:
    return z
  elif y>=x and y>=z:
    return y
  else:
    return x
print("largest number is: ",fun(-10,12,-15))
```

[3]. Write a function to print prime numbers between provided range.

```python
a=10
b=23
def fun(start,end):
  flag=True
  ans=[]
  for i in range(start,end+1):
    flag=True
    for j in range(2,i):
      if i%j==0:
        flag=False
        break
    if flag:
      ans.append(i)
  return ans
print("Prime numbers between ",a," and", b," are: ",fun(a,b))
```

[4]. Write a function to swap two values provided as arguments.

```python
x=10
y=20
def swap(a,b):
  return b,a
print("Before swap: ","x is ",x," and y is ",y)
x,y=swap(x,y)
print("After swap: ","x is ",x," and y is ",y)
```

if rate of interest is not provided consider it as 3.5%.

```python
n=5
p=100000
def interest(n,p,r=3.5):
  return (p*r*n/100)
print("Interest is: ",interest(n,p))
```

[6]. Write a function to compute the sum of data of variable length.

```python
def sum(*args):
    result = 0

    for arg in args:
        result = result + arg

    return result

print('Sum of the numbers is ',sum(10, 20))            
print('Sum of the numbers is ',sum(10, 20, 30))        
print('Sum of the numbers is ',sum(10, 20, 30, 40))    
```
